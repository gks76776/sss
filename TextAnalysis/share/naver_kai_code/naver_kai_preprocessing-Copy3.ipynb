```python
## 문장들을 꺼내서 오브젝트에서 str 타입으로 바꾸는 작업
def str_token_join(data):
    hinix_text_str = ' '
    hinix_str = []
    result = []
    for text_list in data:        
        for text in text_list:
            #마디를 문자열로 변경 
            str_text = str(text)
            # +가 있는 마디가 있을수 있기 split으로 list를 담아줌  
            corpus = str_text[str_text.find('\t')+1:].split(' + ') # 자른걸 원소로 가지도록             
            dict_tags = {'NNG':1,'NNP':1,'NNB':1,'VV':1,'VA':1,'VCP':1,'VCN':1,'MM':1,'MAG':1,'IC':1,'JX':1,'SF':1,'SE':1, 'SW':1, 'SN':1,'XPN':1}#,'SL':1
            
            # 각 list가 원하는 태그인가 아닌가 검증            
            for corpus_element in corpus:
                tag = corpus_element[corpus_element.find('/')+1:]
                tmp = dict_tags.get(tag)                
                if tmp !=None:                   
                    # 이 text를 text_list가 되도록 리스트에 담는다.
                    hinix_str.append(corpus_element)               
        # 리스트가 아닌 str을 담아주기 위해 join을 이용한다.
        result.append(hinix_text_str.join(hinix_str))
        #hinix_str 초기화 
        hinix_str=[]         
    return result
```

```python
start_time=time.time()
hinix_text = hinix['text']#원하는 프로그램 => 1000개씩 잘라서 tokenizer 한다음에 저장한다.
stride = 5000
# 코드 제대로 설계를 못하고 짬.
for _ in tqdm_notebook(range(len(hinix_text)//stride+1)):        
    # 1000개로 쪼개기
    hinix_tmp = hinix_text[:min(stride,len(hinix_text))] #1000개 이하개 일수도 있으므로    
    #tokenizer 하기
    hinix_text_token = hinix_tmp.apply(lambda x:api.analyze(x)) # 1000개 단위로 자른거 형태소 분석 
    hinix_text_token = pd.DataFrame(str_token_join(hinix_text_token))
    #저장하기
    with open('./naver_{0}_token.csv'.format(stock_code),'a',encoding='utf-8') as f:
        hinix_text_token.to_csv(f, header=False, index= False,encoding='utf-8')    
    if len(hinix_text) < stride:
        break    
    hinix_text = hinix_text[stride:]    
print('걸린시간:',time.time()-start_time)
```
