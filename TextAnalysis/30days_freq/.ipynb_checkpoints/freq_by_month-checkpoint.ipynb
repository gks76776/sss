```python
def month_vouch(m):
    if len(str(m))==1:
        return '0'+str(m)
    else:
        return str(m)
```

```python
def pearson_cor(a,b):   
    if a.shape != b.shape:
        print('크기가 다른 데이터입니다.')            
    else:
        return((np.dot(a,b) - np.mean(a)*np.mean(b)*a.shape[0])/(a.shape[0]-1))/(np.std(a)*np.std(b))
```

```python
file_list = glob('./data/freq2/naver*')

# 결과가 이상한 코드 : 051900
# 수익률 자료에 없는 코드 : 282330,298040

file_name_list = [file[file.find('\\')+1:] for file in file_list]    
code_list = [file_name.split("_")[1]for file_name in file_name_list]
all_dict = {}
tag_dict = Counter()
model_result = Counter()
for stock_code in code_list:
    print(stock_code)
    path = './data/freq2/naver_{0}_freq.parquet/'.format(stock_code)
    file_name = os.listdir(path)[0]
    data = pd.read_parquet(path+file_name)
    #data['date'] = pd.to_datetime(data['date'], format='%Y.%m.%d', errors='ignore')

    

    

    #월별 집계
    month_sum = data
    
    month_sum = month_sum.iloc[:-1,:]
    
    prove_frame = month_sum.sum(axis=0)    

    month_sum = month_sum.loc[:,prove_frame>100] # 총 기간 동안 나온 빈도의 수가 100개 이상인 것만 추출

    import numpy as np

    b = np.random.normal(size=data.shape[0])

    #월별 수익률
    earning_month_raw = pd.read_excel('./data/기업별_1개월_수익률.xlsx')

    symbol = earning_month_raw.loc['Symbol Name':,:]
    symbol.columns = earning_month_raw.loc['Symbol']

    # 해당 열 추출
    data = symbol['A'+stock_code]

    earning_data = data[6:]
    earning_data_month_del = [str((date-relativedelta(months=1)).year)+'-'+str(month_vouch(  (date-relativedelta(months=1)).month  )) for date in earning_data.index]
    
    month_del_set = set(earning_data_month_del)
    month_sum_index = month_sum.index
    month_sum_index_set = set(month_sum_index)
    minus_index = list(month_del_set.difference(month_sum_index_set))
    if minus_index != []:
        drop_index = [earning_data_month_del.index(idx) for idx in minus_index]
        earning_data.drop(earning_data.index[drop_index],axis=0,inplace=True)
    
    

    pearson_cor(earning_data,month_sum.iloc[:,1])
    corr_dict = {}
    for idx in range(month_sum.shape[1]):
        tag = month_sum.iloc[:,idx]
        if tag.name == './sf':
            continue
        corr = pearson_cor(np.array(tag),np.array(earning_data))
        #print(type(tag),type(corr))
        tag_dict[tag.name] += 1
        corr_dict[tag.name] = corr
    import operator
    all_dict[stock_code] = sorted(corr_dict.items(), key=operator.itemgetter(1),reverse=True)
    
    # x축 월별 형태소 하나 빈도 , y축 수익률
    
#     for idx in range(month_sum.shape[1]):
        
#         plt.scatter(month_sum.iloc[:,idx],earning_data)
#         plt.title(month_sum.iloc[:,idx].name)
#         plt.show()
#         if idx==10:
#             break
    #plt.plot(month_sum.iloc[:,idx],earning_data)
    one_code_list = all_dict[stock_code]
    tag_list = [tag[0] for tag in one_code_list[:int(len(one_code_list)*0.3)]]
    y = np.where(earning_data>=0,1,0)
    x = month_sum[tag_list]
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, shuffle=False)
    lenx=X_train.shape
    if lenx[0]<=10 or lenx[1]<=2 :
        continue
#     model1 = SVC()
#     model1.fit(X_train,y_train)
#     pre = model1.predict(X_test)
#     ac_score1 = metrics.accuracy_score(y_test, pre)
#     '%0.2f' % ac_score1
    
    
#     model2 = RandomForestClassifier(n_estimators=1000, random_state=100)
#     model2.fit(X_train,y_train)
#     pre = model2.predict(X_test)
#     ac_score2 = metrics.accuracy_score(y_test, pre)
#     '%0.2f' % ac_score2
    
    
#     model3 = 
#     model3.fit(X_train,y_train)
#     pre = model3.predict(X_test)
#     ac_score3 = metrics.accuracy_score(y_test, pre)
#     '%0.2f' % ac_score3
    
#     model4 = XGBClassifier(silent=True, learning_rate=0.01, subsample=0.8,objective='binary:logistic',n_estimators=1000,reg_alpha=0.3)
#     model4.fit(X_train,y_train)
#     pre = model4.predict(X_test)
#     ac_score4 = metrics.accuracy_score(y_test, pre)
#     '%0.2f' % ac_score4
    random_state = 2
    classifiers = []
    classifiers.append(SVC(random_state=random_state))
    
    classifiers.append(AdaBoostClassifier(DecisionTreeClassifier(random_state=random_state),random_state=random_state,learning_rate=0.1))
    classifiers.append(RandomForestClassifier(random_state=random_state,n_estimators=1000,max_depth=4))
    classifiers.append(ExtraTreesClassifier(random_state=random_state))
    classifiers.append(GradientBoostingClassifier(random_state=random_state))
    classifiers.append(MLPClassifier(random_state=random_state))
    classifiers.append(KNeighborsClassifier())
    classifiers.append(LogisticRegression(random_state = random_state))
    classifiers.append(LinearDiscriminantAnalysis())
    classifiers.append(GaussianNB())
    accuracy_list = []
    for models in classifiers:
        models.fit(X_train,y_train)
        pre = models.predict(X_test)
        accuracy_list.append(metrics.accuracy_score(y_test, pre))
    model_result[stock_code]=accuracy_list
    break
    
    
```

```python
def pearson_cor(a,b):
#     a = np.array(a)
#     b = np.array(b)
    if a.shape != b.shape:
        print('크기가 다른 데이터입니다.')
    else:
        return((np.dot(a,b) - np.mean(a)*np.mean(b)*a.shape[0])/(a.shape[0]-1))/(np.std(a)*np.std(b))
```
