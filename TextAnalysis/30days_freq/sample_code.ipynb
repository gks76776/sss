```python
#######################################################################################################################################
stime = time.time()
stride = 6
freq_plate = pd.DataFrame(columns=sample_arrange.columns)
for idx in tqdm_notebook(range(sample_arrange.shape[0]-6)):# 이날짜까지 계산함
    #1열 9.07  7.09
    #30개씩 끊는건 유효하나 30일을 초과해버리면 그 이후 데이터는 사용하지 않는다. for문을 돌리면 속도가 느리다.
    # 처음에는 마지막거만 비교해본다. 비교결과 30 이상 차이가 난다면 앞에꺼부터 비교해서 30일 이하 인것만 더한다.
    target_idx = None 
    last_idx = min(sample_arrange.shape[0],idx+stride)
    start_date = sample_arrange.iloc[idx,0]    
    end_date = sample_arrange.iloc[last_idx,0]
    
    if (end_date - start_date).days >30:
        for next_idx in range(1,stride):
            diff = sample_arrange.iloc[next_idx,1:] - start_date
            if diff.days > 30:
                target_idx = idx+next_idx
    else:
        target_index = last_idx
    # sample_arrange
    sum_freq = sample_arrange.iloc[idx:last_idx,1:].sum().astype(np.int8)
    sum_freq['date'] ='-'.join([str(sample_arrange.iloc[idx,0])[:10],str(6)])    
    sum_freq_sort = sum_freq[sample_arrange.columns]        
    freq_plate.loc[idx]=sum_freq_sort # 인덱스로 인식한다. #loc이 행 추가도 한다.
    # 이를 데이터 프레임에 담아서 출력한다. 


#######################################################################################################################################
#freq_plate = freq_plate.astype(np.int8)
print(time.time()-stime)
```
