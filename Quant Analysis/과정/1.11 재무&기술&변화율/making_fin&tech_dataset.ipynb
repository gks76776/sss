```python
def revenue_Data_reshape(data_revenue):   ## 11월 수익률은 6월 데이터가 없기 떄문에 쓸 수 없다.
    com_name = list(data_revenue.iloc[0][5:])
    #     col_name = '수익률 (3개월)(%)'
    index_list = data_revenue['Item'][1:]

    values = data_revenue.iloc[1:,5:]
    values.columns = com_name

    ## 11월 수익률은 6월 데이터가 없기 떄문에 쓸 수 없다.
    b = values
    b.index = index_list

    b.loc[20140515.0] = np.nan
    b.loc[20150515.0] = np.nan
    b.loc[20160515.0] = np.nan
    b.loc[20170515.0] = np.nan
    b.loc[20180515.0] = np.nan

    ind = list(b.index)
    ind.sort()
    
    return b.loc[ind].iloc[1:]
```

```python
# 회사별로 reshape  20
def reshape_fin_dataset(data):
    ##  쓸데없는 칼럼 제거
    company_name = list((data.iloc[8]))
    col_name = list(data.iloc[11])
    data = data.iloc[13:]
    data.columns = col_name
    col_num = len(list(set(col_name)))
    
    # 더미 데이터셋 생성
    sets = pd.DataFrame()    
    for i in range(len(set(company_name))):
        ## 회사 별로 뽑아오기
        sub_dataset = data.iloc[:,i*col_num : (i+1)*col_num][:-3]
        sub_dataset['Name'] = company_name[i*col_num]
        sets = sets.append(sub_dataset)
        
    return sets
```

```python
def reshape_tech_dataset(data_tech):
    col_name = list(data_tech.Symbol[1:])  # set 노노
    com_name = list(data_tech.iloc[0,5:]) ## set 완료
    values = data_tech.iloc[1:,5:]
    values.columns = com_name
    index = list(set(data_tech['Item']))[1:]
    index.sort()

    full_dataset = pd.DataFrame()

    for i in com_name:
        sub_dataset = values[i]
        a = np.array(sub_dataset)
        sub_dataset = pd.DataFrame(a.reshape(-1,int(len(a)/len(set(col_name))))).T
        sub_dataset['Name'] = i
        sub_dataset.index = index
        sub_dataset = sub_dataset.iloc[:-1].loc[[20130815.0, 20131115.0, 20140215.0, 20140515.0, 20140815.0, 20141115.0, 
                                                 20150215.0, 20150515.0, 20150815.0, 20151115.0, 20160215.0, 20160515.0,
                                                 20160815.0, 20161115.0, 20170215.0, 20170515.0, 20170815.0, 20171115.0, 
                                                 20180215.0, 20180515.0, 20180815.0]]

        full_dataset = full_dataset.append(sub_dataset)

    col = ['변동성 (20일)','거래량 (20일 평균)(주)','외국인보유비중(보통+우선)(%)','공매도거래량 (20일)(주)','순매수수량(개인)(20일합산)(주)',
           '순매수수량(기관계)(20일합산)(주)','순매수수량(외국인계)(20일합산)(주)','PER(IFRS-연결)','PBR(IFRS-연결)','PSR(IFRS-연결)',
           'Name']

    full_dataset.columns = col    
    return full_dataset
```

```python
def merge_data(fin, tech, revenue, stop):
    merged_dataset = pd.DataFrame()
    com_name = fin['Name']
    set_name = list(set(com_name))
    date_index = pd.DatetimeIndex(['2013-11-15', '2014-02-15', '2014-05-15', '2014-08-15', '2014-11-15',
                                   '2015-02-15', '2015-05-15', '2015-08-15', '2015-11-15',
                                   '2016-02-15', '2016-05-15', '2016-08-15', '2016-11-15',
                                   '2017-02-15', '2017-05-15', '2017-08-15', '2017-11-15',
                                   '2018-02-15', '2018-05-15', '2018-08-15'])
    
    for i in list((set_name)):
        # 회사 1개 불러오기
        fin_sub_set = fin[fin.Name == i]
        revenue_sub_set = revenue[i]
        ## 2016년 2분기 데이터도 포함됐기에 뺀다.
        tech_sub_set = tech[tech.Name == i].iloc[1:]
        tech_sub_set.drop(['Name'],axis=1,inplace=True)
        stop_sub_set = stop[i]
        
        revenue_sub_set = revenue_sub_set.rename('수익률 (3개월)(%)')
        stop_sub_set = stop_sub_set.rename('거래정지구분')
        
        fin_sub_set.index = date_index
        revenue_sub_set.index = date_index   
        tech_sub_set.index = date_index
        stop_sub_set.index = date_index

        ## tech & fin & revenue data 병합
        merged_Data = pd.concat([fin_sub_set, tech_sub_set, revenue_sub_set, stop_sub_set],axis=1)

        merged_dataset = merged_dataset.append(merged_Data)
        
    ## 중복되는 칼럼 삭제 요함
    merged_dataset = merged_dataset.loc[:,merged_dataset.columns.duplicated()==False]
    
    ## 내가 보기 편하게 칼럼 재조정
    merged_dataset = merged_dataset[['부채비율(%)', '매출액증가율(전년동기)(%)', '매출총이익증가율(전년동기)(%)',
       '영업이익증가율(전년동기)(%)', '당기순이익증가율(전년동기)(%)', '총포괄이익증가율(전년동기)(%)',
       'EBITDA증가율(전년동기)(%)', '총자산증가율(전년동기)(%)', '총부채증가율(전년동기)(%)',
       '총자본증가율(전년동기)(%)', 'EPS증가율(전년동기)(%)', 'BPS증가율(전년동기)(%)',
       'ROA(총포괄이익)(%)', 'ROE(당기순이익)(%)', '총자산회전율(회)','현금및현금성자산구성비율(%)', '차입부채/EBITDA(배)', 'EBITDA(천원)',
       'CFPS(원)','P/EBITDAPS(배)','변동성 (20일)', '거래량 (20일 평균)(주)', '외국인보유비중(보통+우선)(%)', '공매도거래량 (20일)(주)',
       '순매수수량(개인)(20일합산)(주)', '순매수수량(기관계)(20일합산)(주)', '순매수수량(외국인계)(20일합산)(주)',
       'PER(IFRS-연결)', 'PBR(IFRS-연결)', 'PSR(IFRS-연결)','수익률 (3개월)(%)', 'Name','거래정지구분']]
    
    delete_com = ['현대그린푸드']
    
    for i in delete_com:
        merged_dataset = merged_dataset[merged_dataset.Name != i]
    
    # 수익률 칼럼 데이터 타입 변경
    merged_dataset['수익률 (3개월)(%)'] = merged_dataset['수익률 (3개월)(%)'].astype('float64')

    return merged_dataset
```

```python
def making_rate_of_change(sub_dataset,col):
    # 변화율 칼럼 삽입 / 기존 칼럼 삭제*********************************************************************
    for colname in col:
        # 변화율 백터 생성
        a = sub_dataset[colname]
        g = list(a.pct_change())

        ## 분보가 - 일때 방향성이 달라져서 -1 곱하기
        for i in list(range(len(a)))[:-1]:
            if list(a)[i] < 0:
                g[i+1] = g[i+1] * (-1)
                
            elif list(a)[i] == 0:
                g[i+1] = np.nan
                    
        sub_dataset[colname+'변화율'] = g
    
    sub_dataset = sub_dataset[1:]
        
    return sub_dataset
```

```python
def make_cate_t(data):
    ## 수익률 범주화
    bb = np.array([-100,0,100])

    Y_data_cat = np.digitize(data['수익률 (3개월)(%)'],bb)
    Y_data_cat = Y_data_cat - 1
    
    return Y_data_cat
```

```python
def make_cate_f(data):
    ## 수익률 범주화
    bb = np.array([-100,-15,0,15,100])

    Y_data_cat = np.digitize(data['수익률 (3개월)(%)'],bb)
    Y_data_cat = Y_data_cat - 1
    
    return Y_data_cat
```
