```python
def revenue_Data_reshape(Data):   ## 11월 수익률은 6월 데이터가 없기 떄문에 쓸 수 없다.
    com_name = list(Data.iloc[0][5:])
#     col_name = '수익률 (3개월)(%)'
    index_list = Data['Item'][2:]
    
    values = Data.iloc[1:,5:]
    values.columns = com_name
    
    ## 11월 수익률은 6월 데이터가 없기 떄문에 쓸 수 없다.
    b = values[1:]
    b.reset_index(drop=True,inplace = True)
    b.index = index_list
    
    b.loc[20140515.0] = np.nan
    b.loc[20150515.0] = np.nan
    b.loc[20160515.0] = np.nan
    b.loc[20170515.0] = np.nan
    b.loc[20180515.0] = np.nan
    
    ind = list(b.index)
    ind.sort()
    
    return b.loc[ind]
```

```python
# 회사별로 reshape  20
def reshape_fin_dataset(data):
    ##  쓸데없는 칼럼 제거
    company_name = list((data.iloc[8]))
    col_name = list(data.iloc[11])
    data = data.iloc[13:-2]
    data.columns = col_name
    col_num = len(list(set(col_name)))
    
    # 더미 데이터셋 생성
    sets = pd.DataFrame()    
    for i in range(len(set(company_name))):
        ## 회사 별로 뽑아오기
        sub_dataset = data.iloc[:,i*col_num : (i+1)*col_num]
        sub_dataset['Name'] = company_name[i*col_num]
        sets = sets.append(sub_dataset)
    sets.drop(['수익률 (3개월)(%)'],inplace=True,axis=1)
    return sets
```

```python
def reshape_tech_dataset(data_tech):
    col_name = list(data_tech.Symbol[1:])  # set 노노
    com_name = list(data_tech.iloc[0,5:]) ## set 완료
    day = data_tech.Item[1:]
    values = data_tech.iloc[1:,2:]
    values = values.drop(['Currency'],axis=1)
    values.columns = ['Item','Symbol'] + com_name
    values.reset_index(drop=True, inplace=True)
    index = list(set(data_tech['Item']))[1:]
    index.sort()

    full_dataset = pd.DataFrame()

    for i in com_name:
        sub_dataset = values[i]
        a = np.array(sub_dataset)
        sub_dataset = pd.DataFrame(a.reshape(-1,int(len(a)/len(set(col_name))))).T
        sub_dataset['Name'] = i
        sub_dataset.index = index
        sub_dataset = sub_dataset.iloc[:-1]

        full_dataset = full_dataset.append(sub_dataset)

    col = ['변동성 (20일)','거래량 (20일 평균)(주)','외국인보유비중(티커)(%)','공매도거래량 (20일)(주)','순매수수량(개인)(20일합산)(주)',
           '순매수수량(기관계)(20일합산)(주)','순매수수량(외국인계)(20일합산)(주)','PER(IFRS-연결)','PSR(IFRS-연결)','PBR(IFRS-연결)',
           '수익률 (1개월)(%)','거래정지구분','Name']

    full_dataset.columns = col
    full_dataset.reset_index(drop=True,inplace=True)
    
    return full_dataset
```

```python
def merge_data(fin,tech,revenue):
    merged_dataset = pd.DataFrame()
    com_name = fin['Name']
    set_name = list(set(com_name))
    index = list(set(fin.index))
    index.sort()
    
    for i in list(range(len(set_name))):
        # 회사 1개 불러오기
        fin_sub_set = fin[ fin.Name == set_name[i]]
        tech_sub_set = tech[ tech.Name == set_name[i]]
        revenue_sub_set = pd.DataFrame(list(revenue[set_name[i]]),columns=['수익률 (3개월)(%)'])


        fin_sub_set.index = index
        tech_sub_set.index = index
        revenue_sub_set.index = index   

        ## tech & fin & revenue data 병합
        merged_Data = pd.concat([fin_sub_set,tech_sub_set,revenue_sub_set],axis=1)

        merged_dataset = merged_dataset.append(merged_Data)
        
    ## 중복되는 칼럼 삭제 요함
    merged_dataset = merged_dataset.loc[:,merged_dataset.columns.duplicated()==False]
    
    ## 내가 보기 편하게 칼럼 재조정
    merged_dataset = merged_dataset[['부채비율(%)', '매출액증가율(전년동기)(%)','매출액증가율(전분기)(%)',
   '매출총이익증가율(전년동기)(%)', '매출총이익증가율(전분기)(%)', '영업이익증가율(전년동기)(%)',
   '영업이익증가율(전분기)(%)', '당기순이익증가율(전년동기)(%)', '당기순이익증가율(전분기)(%)',
   '총포괄이익증가율(전년동기)(%)', '총포괄이익증가율(전분기)(%)', 'EBITDA증가율(전년동기)(%)',
   'EBITDA증가율(전분기)(%)', '총자산증가율(전년동기)(%)', '총자산증가율(전분기)(%)',
   '총부채증가율(전년동기)(%)', '총부채증가율(전분기)(%)', '총자본증가율(전년동기)(%)',
   '총자본증가율(전분기)(%)', 'EPS증가율(전년동기)(%)', 'EPS증가율(전분기)(%)',
   'BPS증가율(전년동기)(%)', 'BPS증가율(전분기)(%)', 'ROA(총포괄이익)(%)', 'ROE(당기순이익)(%)',
   '총자산회전율(회)', '총부채회전율(회)', '총자본회전율(회)', '현금및현금성자산구성비율(%)',
   '차입부채/EBITDA(배)', '변동성 (20일)', '거래량 (20일 평균)(주)',
   '외국인보유비중(티커)(%)', '공매도거래량 (20일)(주)', '순매수수량(개인)(20일합산)(주)',
   '순매수수량(기관계)(20일합산)(주)', '순매수수량(외국인계)(20일합산)(주)', 'PER(IFRS-연결)',
   'PSR(IFRS-연결)', 'PBR(IFRS-연결)', '수익률 (1개월)(%)','수익률 (3개월)(%)', '거래정지구분', 'Name']]
    
    delete_com = ['우리은행', 'DB손해보험', 'BNK금융지주','한미약품', '삼성생명', '한화생명', 'KB금융',
                  '삼성카드', '하나금융지주', '셀트리온', '한국금융지주', '신한지주', '한국가스공사', '메리츠종금증권',
                  '현대해상', '한국전력', '녹십자', '부광약품', '삼성증권', '녹십자홀딩스', 'JW중외제약', 'NH투자증권',
                  '미래에셋대우', '일양약품']
    for i in delete_com:
        merged_dataset = merged_dataset[merged_dataset.Name != i]
    
    # 수익률 칼럼 데이터 타입 변경
    merged_dataset['수익률 (1개월)(%)'] = merged_dataset['수익률 (1개월)(%)'].astype('float64')
    merged_dataset['수익률 (3개월)(%)'] = merged_dataset['수익률 (3개월)(%)'].astype('float64')

    return merged_dataset
```

```python
def making_rate_of_change(sub_dataset,col):
    # 변화율 칼럼 삽입 / 기존 칼럼 삭제*********************************************************************
    for colname in col:
        # 변화율 백터 생성
        a = sub_dataset[colname]
        g = list(a.pct_change())

        ## 분보가 - 일때 방향성이 달라져서 -1 곱하기
        for i in list(range(len(a)))[:-1]:
            if list(a)[i] < 0:
                g[i+1] = g[i+1] * (-1)
                
            elif list(a)[i] == 0:
                g[i+1] = np.nan
                    
        sub_dataset[colname] = g
    
    sub_dataset = sub_dataset[1:]
        
    return sub_dataset
```

```python
def make_cate_three(data):
    ## 수익률 범주화
    bb = np.array([-100,0,100])

    Y_data_cat_three = np.digitize(data['수익률 (3개월)(%)'],bb)
    Y_data_cat_three = Y_data_cat_three - 1
    
    Y_data_cat_three
    
    return Y_data_cat_three
```

```python
def make_cate_f(data):
    ## 수익률 범주화
    bb = np.array([-100,-15,0,15,100])

    Y_data_cat_three = np.digitize(data['수익률 (3개월)(%)'],bb)
    Y_data_cat_three = Y_data_cat_three - 1
    
    Y_data_cat_three
    
    return Y_data_cat_three
```

```python
def std_based_outlier(df):
    cols = ['매출액증가율(전년동기)(%)','매출액증가율(전분기)(%)',
           '매출총이익증가율(전년동기)(%)', '매출총이익증가율(전분기)(%)', '영업이익증가율(전년동기)(%)',
           '영업이익증가율(전분기)(%)', '당기순이익증가율(전년동기)(%)', '당기순이익증가율(전분기)(%)',
           '총포괄이익증가율(전년동기)(%)', '총포괄이익증가율(전분기)(%)', 'EBITDA증가율(전년동기)(%)',
           'EBITDA증가율(전분기)(%)', '총자산증가율(전년동기)(%)', '총자산증가율(전분기)(%)',
           '총부채증가율(전년동기)(%)', '총부채증가율(전분기)(%)', '총자본증가율(전년동기)(%)',
           '총자본증가율(전분기)(%)', 'EPS증가율(전년동기)(%)', 'EPS증가율(전분기)(%)',
           'BPS증가율(전년동기)(%)', 'BPS증가율(전분기)(%)']
    
    for i in range(0, len(cols)):
        df.iloc[:,i] = df.iloc[:,i].replace(0, np.NaN)
        df = df[~(np.abs(df.iloc[:,i] - df.iloc[:,i].mean()) > (3*df.iloc[:,i].std()))].fillna(0)
    
    return(df)
```

```python
## 증가율로 바꾸기 / null값 지우기 등.. 실행
def preprocessing_data(data): 
    # 쓸 변수 선택
    data = data[['매출액증가율(전년동기)(%)','매출액증가율(전분기)(%)',
                   '매출총이익증가율(전년동기)(%)', '매출총이익증가율(전분기)(%)', '영업이익증가율(전년동기)(%)',
                   '영업이익증가율(전분기)(%)', '당기순이익증가율(전년동기)(%)', '당기순이익증가율(전분기)(%)',
                   '총포괄이익증가율(전년동기)(%)', '총포괄이익증가율(전분기)(%)', 'EBITDA증가율(전년동기)(%)',
                   'EBITDA증가율(전분기)(%)', '총자산증가율(전년동기)(%)', '총자산증가율(전분기)(%)',
                   '총부채증가율(전년동기)(%)', '총부채증가율(전분기)(%)', '총자본증가율(전년동기)(%)',
                   '총자본증가율(전분기)(%)', 'EPS증가율(전년동기)(%)', 'EPS증가율(전분기)(%)',
                   'BPS증가율(전년동기)(%)', 'BPS증가율(전분기)(%)','거래정지구분','수익률 (3개월)(%)','3개월 수익률(f)','Name']]
    
    col_name = data.columns
    company_name = data.Name
    scaled_total_dataset = pd.DataFrame()

    for name in list(set(company_name)):
        
        ## 회사 별로 뽑아오기
        sub_dataset = data[data.Name == name]
        
        #거래정지구분에서 null 값있는 row 제거(상장 전 기간 제거)
        select_notNan = sub_dataset['거래정지구분'].dropna()
        sub_dataset = sub_dataset.loc[select_notNan.index]
        
        # 거래정지 로우 & 칼럼 제거
        sub_dataset = sub_dataset[sub_dataset['거래정지구분'] !='거래정지']
        sub_dataset.drop(['거래정지구분'],inplace=True,axis=1)
        
#         #결측치 처리 구간 ***********************************************************************
#         per, psr, pbr => null은 칼럼 평균으로 대체 
#         sub_dataset['PER(IFRS-연결)'].fillna(sub_dataset['PER(IFRS-연결)'].mean(),inplace = True)
#         sub_dataset['PSR(IFRS-연결)'].fillna(sub_dataset['PSR(IFRS-연결)'].mean(),inplace = True)
#         sub_dataset['PBR(IFRS-연결)'].fillna(sub_dataset['PBR(IFRS-연결)'].mean(),inplace = True)
        
#         # 변화율로 만들기
#         sub_dataset = making_rate_of_change(sub_dataset,['부채비율(%)','ROA(총포괄이익)(%)', 'ROE(당기순이익)(%)',
#        '총자산회전율(회)', '총부채회전율(회)', '총자본회전율(회)','현금및현금성자산구성비율(%)', '변동성 (20일)','외국인보유비중(티커)(%)',
#        'PER(IFRS-연결)','PSR(IFRS-연결)', 'PBR(IFRS-연결)'])            
        
        ## 수익률 없는 칼럼 삭제
        sub_dataset = sub_dataset[sub_dataset['수익률 (3개월)(%)'].isna() == False]
        
            ## inf 처리
        for i in sub_dataset.columns:
            sub_dataset = sub_dataset.loc[sub_dataset[i] != np.inf]
        
        ### 이상치 처리 ## ************************************************************************************        
        sub_dataset = std_based_outlier(sub_dataset)
        
        #스케일링
#         sub_dataset = scaling_data(sub_dataset)
        
        scaled_total_dataset = scaled_total_dataset.append(sub_dataset)

    
    scaled_total_dataset = scaled_total_dataset[scaled_total_dataset['수익률 (3개월)(%)'] < 60]
    scaled_total_dataset = scaled_total_dataset[scaled_total_dataset['수익률 (3개월)(%)'] > -60 ]

    ### 수익률 변환
    scaled_total_dataset['3개월 수익률(f)'] = make_cate_f(scaled_total_dataset)
    scaled_total_dataset['수익률 (3개월)(%)'] = make_cate_three(scaled_total_dataset) 

    
    
    
    ## 결측치 있는 행 삭제
    scaled_total_dataset.dropna(axis=0,inplace=True)

    return scaled_total_dataset
    
```
