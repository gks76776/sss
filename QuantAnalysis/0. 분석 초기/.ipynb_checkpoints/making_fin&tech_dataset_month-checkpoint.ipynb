```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
```

```python
data_tech = pd.read_excel('./data/kospi200_공시기간_조정_전체_데이터_2.xlsx',sheet_name='1개월단위_tech')
```

```python
# 회사별로 reshape  20
def reshape_tech_dataset(data_tech):
    ##  쓸데없는 칼럼 제거
    company_name = list((data_tech.iloc[8]))
    col_name = list(data_tech.iloc[11])
    data_tech = data_tech.iloc[13:]
    data_tech.columns = col_name
    col_num = len(list(set(col_name)))
    sets = pd.DataFrame()
    dummy_rate = pd.DataFrame([np.nan])
    for i in range(len(set(company_name))):

        ## 회사 별로 뽑아오기
        sub_dataset = data_tech.iloc[:,i*col_num : (i+1)*col_num]
        sub_dataset['Name'] = company_name[i*col_num]
        revenue = sub_dataset.pop('수익률 (1개월)(%)')

        index_list = revenue.index
        ## 수익률 3개월 1행씩 땡기기
        b = revenue[1:]
        b = b.append(dummy_rate)
        b.index = index_list
        b.columns = ['수익률 (1개월)(%)']
        sub = pd.concat([sub_dataset,b], axis=1)
        sub_Se = sub[pd.notna(sub['수익률 (1개월)(%)'])][1:]
        
        sets = sets.append(sub_Se)
    return sets
```

```python
tech = reshape_tech_dataset(data_tech)
tech
```

```python
def making_rate_of_change(sub_dataset,col):
    # 변화율 칼럼 삽입 / 기존 칼럼 삭제*********************************************************************
    for colname in col:
        
        # 변화율 백터 생성
        a = sub_dataset[colname]
        g = list(a.pct_change())

        ## 분보가 - 일때 방향성이 달라져서 -1 곱하기
        for i in list(range(len(a)))[:-1]:
            if list(a)[i] < 0:
                g[i+1] = g[i+1] * (-1)
                
            elif list(a)[i] == 0:
                g[i+1] = np.nan
                    
        sub_dataset[colname] = g
    
    sub_dataset = sub_dataset[1:]
        
    return sub_dataset
```

```python
def scaling_data(sub_set):
    scaled_data = pd.DataFrame()

    sub_set = sub_set.iloc[1:]
    sub_set_y = sub_set.index
    
    scale_col = ['변동성 (20일)', '거래량 (20일 평균)(주)', '외국인보유비중(보통+우선)(%)', '공매도거래량 (20일)(주)',
       '순매수수량(개인)(20일합산)(주)', '순매수수량(기관계)(20일합산)(주)', '순매수수량(외국인계)(20일합산)(주)',
                'PER(IFRS-연결)', 'PBR(IFRS-연결)', 'PSR(IFRS-연결)']
    unscale_col = ['Name','수익률 (1개월)(%)']
    
    sub_set_x = sub_set[scale_col]
    sub_set_no = sub_set[unscale_col]
    
    ## 스케일 부분
    scaled_sub_set_x = pd.DataFrame(scale(sub_set_x),index = sub_set_y ,columns = scale_col)
    x = pd.concat([scaled_sub_set_x,sub_set_no],axis = 1)
    
    return x
```

```python
def make_cate_three(data,bound1):
    ## 수익률 범주화
    bb = np.array([-100,bound1,100])

    Y_data_cat_three = np.digitize(data['수익률 (1개월)(%)'],bb)
    Y_data_cat_three = Y_data_cat_three - 1
    
    data['수익률 (1개월)(%)'] = Y_data_cat_three
    
    return data
```

```python
## 증가율로 바꾸기 / null값 지우기 등.. 실행
def preprocessing_data(data):
    col_name = data.columns
    company_name = data.Name
    scaled_total_dataset = pd.DataFrame()

    for name in list(set(company_name)):
        
        ## 회사 별로 뽑아오기
        sub_dataset = data[data.Name == name]
        
        #거래정지구분에서 null 값있는 row 제거(상장 전 기간 제거)
        select_notNan = sub_dataset['거래정지구분'].dropna()
        sub_dataset = sub_dataset.loc[select_notNan.index]
        
        # 거래정지 로우 & 칼럼 제거
        sub_dataset = sub_dataset[sub_dataset['거래정지구분'] !='거래정지']
        sub_dataset.drop(['거래정지구분'],inplace=True,axis=1)
        
        # 결측치 처리 구간 ***********************************************************************
        # per, psr, pbr => null은 칼럼 평균으로 대체 
        sub_dataset['PER(IFRS-연결)'].fillna(sub_dataset['PER(IFRS-연결)'].mean(),inplace = True)
        sub_dataset['PSR(IFRS-연결)'].fillna(sub_dataset['PSR(IFRS-연결)'].mean(),inplace = True)
        sub_dataset['PBR(IFRS-연결)'].fillna(sub_dataset['PBR(IFRS-연결)'].mean(),inplace = True)
        
        # 변화율로 만들기
        sub_dataset = making_rate_of_change(sub_dataset,['변동성 (20일)', '외국인보유비중(보통+우선)(%)',
                                                         'PER(IFRS-연결)', 'PBR(IFRS-연결)', 'PSR(IFRS-연결)'])
        
#             ## inf 처리
#         for i in sub_dataset.columns:
#             sub_dataset = sub_dataset.loc[sub_dataset[i] != np.inf]
        
        ### 이상치 처리 ## ************************************************************************************        
        
        # 스케일링
        sub_dataset = scaling_data(sub_dataset)
        
        scaled_total_dataset = scaled_total_dataset.append(sub_dataset)
    
    ## NA 값 처리  ************************************************************************************
    
    
    
    scaled_total_dataset = scaled_total_dataset[scaled_total_dataset['수익률 (1개월)(%)'] < 50]
    scaled_total_dataset = scaled_total_dataset[scaled_total_dataset['수익률 (1개월)(%)'] > -50 ]
    
    
    scaled_total_dataset['수익률 (1개월)(%)'] = scaled_total_dataset['수익률 (1개월)(%)'].astype('float64')
    ### 수익률 변환
    final_dataset = make_cate_three(scaled_total_dataset,8) 

#     ## 상관관계가 강한 변수 제거
#     final_dataset = final_dataset[['부채비율(%)','매출액증가율(전년동기)(%)', '매출액증가율(전분기)(%)',
#        '당기순이익증가율(전년동기)(%)', '당기순이익증가율(전분기)(%)',
#        '총포괄이익증가율(전년동기)(%)', 'EBITDA증가율(전년동기)(%)',
#        'EBITDA증가율(전분기)(%)', '총자산증가율(전년동기)(%)', '총자산증가율(전분기)(%)',
#        '총자본증가율(전년동기)(%)','총자본증가율(전분기)(%)',
#        'BPS증가율(전년동기)(%)', 'BPS증가율(전분기)(%)', 'ROA(총포괄이익)(%)', 
#        '총자산회전율(회)', '현금및현금성자산구성비율(%)',
#        '차입부채/EBITDA(배)', '변동성 (20일)', '거래량 (20일 평균)(주)', '외국인보유비중(티커)(%)',
#        '공매도거래량 (20일)(주)', '순매수수량(개인)(20일합산)(주)', '순매수수량(기관계)(20일합산)(주)',
#        '순매수수량(외국인계)(20일합산)(주)', 'PER(IFRS-연결)', 'PSR(IFRS-연결)', 'PBR(IFRS-연결)',
#        '수익률 (1개월)(%)','수익률 (3개월)(%)','Name']]
    
    return final_dataset
    
```
